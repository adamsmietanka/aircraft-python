```python
import pandas as pd
import plotly
```

```python
ct = pd.read_csv('raf2-ct.csv', dtype=float)
cp = pd.read_csv('raf2-cp.csv', dtype=float)
eff = pd.read_csv('raf2-eff.csv', dtype=float)
eff
```

```python
ct = ct[ct['x'] >= 0].reset_index(drop=True)
cp = cp[cp['x'] >= 0].reset_index(drop=True)
eff = eff[eff['x'] >= 0].reset_index(drop=True)
ct.at[0, 'x'] = 0
cp.at[0, 'x'] = 0
eff.at[0, 'x'] = 0

ct
```

```python
angle = '15'
ct_from_cp = cp[angle]*eff[angle]/cp['x']
comparison = pd.DataFrame({'j': ct['x'], 'calculated_ct': ct_from_cp, 'read_ct': ct[angle]})
comparison['error'] = (comparison.calculated_ct - comparison.read_ct) / comparison.read_ct * 100
comparison.head(30)
```
